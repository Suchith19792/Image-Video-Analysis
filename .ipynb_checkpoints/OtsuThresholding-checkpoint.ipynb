```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pdb
```

```python
fname = 'julia'
img = cv2.imread('OtsuThresholding/'+fname+'.png',0)
plt.imshow(img,cmap='gray')
```

```python
hist = plt.hist(img.ravel(),256,[0,255],normed=True)
plt.show()
```

```python
#distance between classes
dbc = np.zeros(256)
#within class variance
wcv = np.zeros(256)
for t in range(0,256):
    mu1,v1,mu2,v2=0,0,0,0
    q1 = np.sum(hist[0][:t]) #/sum(hist[0])
    q2 = 1 - q1
    if q1 != 0.0:
        mu1 = 1/q1*np.sum(np.multiply(np.array(range(t+1)),hist[0][:t+1]))
        v1 = 1/q1*np.sum(np.multiply(np.power(np.array(range(t+1))-mu1,2),hist[0][:t+1]))
    if q2 != 0.0:
        mu2 = 1/q2*np.sum(np.multiply(np.array(range(t+1,256)),hist[0][t+1:]))
        v2 = 1/q2*np.sum(np.multiply(np.power(np.array(range(t+1,256))-mu2,2),hist[0][t+1:]))
    #print(q1,q2,mu1,mu2)
    dbc[t] = q1*q2*(mu1-mu2)**2
    wcv[t] = q1*v1 + q2*v2
    
#dbc
```

```python
thresh1 = np.argmax(dbc)
max_dbc = np.max(dbc)
thresh2 = np.argmin(wcv[1:-1])
min_wcv = np.min(wcv[1:-1])
print(thresh1, max_dbc)
print(thresh2, min_wcv)
```

```python
img_out = img > thresh2
```

```python
plt.imshow(img_out,cmap='gray')
plt.title('theshold: '+str(thresh2))
plt.savefig('OtsuThresholding/'+fname+'_otsu.png',bbox_inches='tight')
```
